# Cluster Configurations

This notebook contains the code to create / update the **Cluster Configurations** for use in the various notebooks, incl.:
- software environments used by workers
- memory and CPU configurations
- number of workers
- idle_timeout setting

## Importing Libraries

```python
import coiled
```

## Cluster for Notebook 01-rrp-data-wrangling

```python
%%capture
# create software environment with necessary libraries

path_to_yml = "/Users/richard/Desktop/conda_env_configs/cap3-wrangling.yml"

coiled.create_software_environment(
    name="cap3-wrangling", 
    conda=path_to_yml,
    post_build=["camel_data light"]
)
```

```python
%%capture
# create cluster configuration
coiled.create_cluster_configuration(
    name="cap3-wrangling-s2_8-w2_16",
    scheduler_cpu=2,
    scheduler_memory="8 GiB",
    worker_cpu=2,
    worker_memory="16 GiB",
    software="cap3-wrangling",
    worker_options={'processes': True}
)
```
