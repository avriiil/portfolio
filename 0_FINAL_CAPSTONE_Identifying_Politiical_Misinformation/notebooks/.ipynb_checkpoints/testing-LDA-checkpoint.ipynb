```python
def get_lemmas(tokenized_text):
    disambig = mle.disambiguate(tokenized_text)
    try:
        lemmas = [d.analyses[0].analysis['lex'] for d in disambig]
        return lemmas
    except:
        return np.nan
```

OK. We now have a lemmatized subset of our data.

We can start taking the LDA for a spin.

The approach is as follows:
- Tokenize and Lemmatize
- Create a Bag of Words that counts how many times each word occurs in the training set
   - filter out extremes
- For each document, create dictionary that contains words as indices into BOW and number of times that words appears
   - e.g. word # 432 appears 3 times (432,3)
   - "bow_corpus"
- Create a tf-idf model
- Train an LDA.multicore model on our bow_corpus
- Train an LDA on our tf-idf

Consider adding **n-grams** to aid the model, i.e. groups of words treated as single tokens -- this takes context into account.

Read [this article](https://www.kdnuggets.com/2019/11/topics-extraction-classification-online-chats.html):
- "add n-grams to the dataset to guide the topic model. The assumption here is that short sequences of words treated as single entities, or tokens, usually contain useful information to identify the topics of a sentence. Only bigrams turned out to be meaningful in this context — indeed they significantly improved the topic model performance"

# Classify Tweet Topics (Full Dataset)

```python
# define classifying function
def get_topics(df):
    docs = df.tweet_text.to_numpy()
    df['topic'] = 0
    
    for i in range(len(docs)):
        try:
            df['topic'][i] = sorted(lda_model.get_document_topics(bow_corpus[i], minimum_probability=0.5), reverse=True)[0][0]
        except:
            df['topic'][i] = np.nan
            
    return df
```
